# MATPLOTLIB

Was ist Matplotlib?
- Ein Plotting-Toolkit für Python, das 2D- und einfache 3D-Diagramme erzeugt.
- Hauptmodul heißt pyplot, das eine Schnittstelle ähnlich wie MATLAB bietet.
- Weit verbreitet in Wissenschaft, Technik, Data Science.

Grundlegende Konzepte
- Figure: Das gesamte Bild oder Fenster, das eine oder mehrere Plots enthalten kann.
- Axes: Ein einzelnes Diagramm in der Figure (die eigentliche “Zeichenfläche”).
- Plot: Die grafische Darstellung von Daten (Linien, Punkte, Balken etc.).

Typische Schritte zum Plotten
1.	Daten vorbereiten (z.B. Listen oder NumPy-Arrays).
2.	Figure und Axes erstellen (optional, wenn du mehrere Plots möchtest).
3.	Daten plotten z.B. mit plt.plot().
4.	Achsen beschriften, Titel hinzufügen, Legende anzeigen.
5.	Plot anzeigen mit plt.show().

## 2D Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Einfacher Sinus-Plot')
plt.legend()
plt.show()
```

### 2D Plots nebeneinander (Subplots)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 Reihe, 2 Spalten

# Erster Plot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Zweiter Plot
axs[1].plot(x, y2, color='red')
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.tight_layout()  # damit sich Plots nicht überlappen
plt.show()
```

## 3D Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Gitter erzeugen
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Funktion definieren, z.B. Z = sin(r)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# 3D-Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Oberfläche plotten
ax.plot_surface(X, Y, Z, cmap='viridis')

# Achsentitel
ax.set_title('3D-Oberfläche: sin(r)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

### 3D Plots nebeneinander (Subplots)

```python
import numpy as np
import matplotlib.pyplot as plt

# Gitter erzeugen
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z1 = np.sin(np.sqrt(X**2 + Y**2))
Z2 = np.cos(np.sqrt(X**2 + Y**2))

# Figur und Subplots erstellen
fig = plt.figure(figsize=(12, 5))

# Erster 3D-Plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis')
ax1.set_title('3D Plot: sin(r)')

# Zweiter 3D-Plot
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='plasma')
ax2.set_title('3D Plot: cos(r)')

plt.tight_layout()
plt.show()
```

## CMAPS und Farbskala

```python
import numpy as np
import matplotlib.pyplot as plt

# Beispiel-Daten (2D-Gitter)
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Alle verfügbaren Colormap-Gruppen
cmap_categories = {
    "Perceptually Uniform Sequential": ['viridis', 'plasma', 'inferno', 'magma', 'cividis'],
    "Sequential": ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds'],
    "Sequential (2)": ['YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu', 'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'],
    "Diverging": ['PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu', 'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic'],
    "Cyclic": ['twilight', 'twilight_shifted', 'hsv'],
    "Qualitative": ['Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 'Set1', 'Set2', 'Set3', 'tab10', 'tab20', 'tab20b', 'tab20c'],
    "Miscellaneous": ['flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'gist_stern', 'gnuplot', 'gnuplot2', 'CMRmap', 'cubehelix', 'brg', 'gist_rainbow', 'rainbow', 'jet', 'nipy_spectral', 'gist_ncar']
}

# Plot-Funktion
for category, cmaps in cmap_categories.items():
    n = len(cmaps)
    fig, axs = plt.subplots(1, n, figsize=(3*n, 3))
    fig.suptitle(category, fontsize=16)

    for ax, cmap in zip(axs, cmaps):
        im = ax.imshow(Z, cmap=cmap, extent=(-3, 3, -3, 3))
        ax.set_title(cmap)
        plt.colorbar(im, ax=ax, shrink=0.7)

    plt.tight_layout()
    plt.show()
```
